```python
df_kerala
```

```python
thresold = 2400
flood = []
jun_to_sep = list(df_kerala['JJAS'])
for i in range(0,len(df_kerala)):
    if jun_to_sep[i] > thresold:
        flood.append(1)
    else:
        flood.append(0) 
```

```python
pd.set_option('mode.chained_assignment',None)
df_kerala['FLOOD'] = flood
```

```python
df_kerala
```

```python
#Preprocessing on kerala_df
```

```python
df_kerala.drop(['SUBDIVISION', 'YEAR'], axis=1, inplace=True)
```

```python
df_kerala
```

```python
df_kerala.isna().sum()
```

```python
#scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_kerala.iloc[:,:-1] = scaler.fit_transform(df_kerala.iloc[:,:-1])
```

```python
df_kerala
```

```python
X = df_kerala.drop('FLOOD', axis=1)
y = df_kerala['FLOOD']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=41)
```

```python
#model building | cross validation | hyper parameter tuning
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import LinearSVC, SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV, StratifiedKFold, KFold, cross_val_score
from sklearn.metrics import classification_report, confusion_matrix
```

```python
def get_best_params(X, y):
    models = {
        'knn': {
            'model': KNeighborsClassifier(),
            'params': {
                'n_neighbors': [5,10,15]
            }
        },
        'svc': {
            'model': SVC(gamma='auto'),
            'params': {
                'kernel': ['linear', 'rbf', 'poly'],
                'C': [5,10,15]
            }
        },
        'RF': {
            'model': RandomForestClassifier(),
            'params': {
                'n_estimators': [50,100]
            }
        }
    }
    scores = []
    for model_name, config in models.items():
        cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=41)
        gd = GridSearchCV(estimator=config['model'], param_grid=config['params'], n_jobs=1, cv=cv)
        gd.fit(X,y)
        scores.append({'model': model_name, 'cross_val_score': gd.best_score_, 'best_params': gd.best_params_})
    return pd.DataFrame(scores)
```

```python
get_best_params(X,y)
```

```python
#random forest algorithm performs better compared to other models
```

```python
rf_clf = RandomForestClassifier(n_estimators=50)
rf_clf.fit(X_train, y_train)
y_pred = rf_clf.predict(X_test)

print("Training score:", rf_clf.score(X_train, y_train))
print("Testing score:", rf_clf.score(X_test, y_test))
```

```python
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True)
```

```python
print(classification_report(y_test, y_pred))
```

```python
#lot of outliers present in the data, so we can fill the missing values with the median value.
```

```python
#pipeline for data transformation (it will do the entire preprocessing and return the transformed data)
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

data_pipe = Pipeline([
    ("num_imputer", SimpleImputer(strategy='median')),
    ("scaler", StandardScaler())
])

unlabelled_data.loc[:,:] = data_pipe.fit_transform(unlabelled_data)
trans_df = unlabelled_data
```
